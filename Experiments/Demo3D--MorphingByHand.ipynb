### Play around with a small 3D lattice by changing all its parameters!

```python
from ipywidgets import interact
import ipywidgets as widgets

import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import seaborn as sns
from LattyMorph.morphing import create_flat_3D_sheet

lattice = create_flat_3D_sheet(1,2)
```

```python
def plot_func(cell0_phiz, cell0_phiy, cell0_thetaz, cell0_thetax, cell0_phizP, cell0_phiyP, cell0_thetazP, cell0_thetaxP, cell1_phiz, cell1_phiy, cell1_thetaz, cell1_thetax, cell1_thetazP):
    lattice.config[0][0]['phiz'].data[0] = cell0_phiz*np.pi
    lattice.config[0][0]['phiy'].data[0] = cell0_phiy*np.pi
    lattice.config[0][0]['thetaz'].data[0] = cell0_thetaz*np.pi
    lattice.config[0][0]['thetax'].data[0] = cell0_thetax*np.pi

    lattice.config[0][0]['phizPrime'].data[0] = cell0_phizP*np.pi
    lattice.config[0][0]['phiyPrime'].data[0] = cell0_phiyP*np.pi
    lattice.config[0][0]['thetazPrime'].data[0] = cell0_thetazP*np.pi
    lattice.config[0][0]['thetaxPrime'].data[0] = cell0_thetaxP*np.pi

    lattice.config[0][1]['phiz'].data[0] = cell1_phiz*np.pi
    lattice.config[0][1]['phiy'].data[0] = cell1_phiy*np.pi
    lattice.config[0][1]['thetaz'].data[0] = cell1_thetaz*np.pi
    lattice.config[0][1]['thetax'].data[0] = cell1_thetax*np.pi
    lattice.config[0][1]['thetazPrime'].data[0] = cell1_thetazP*np.pi
    
    check = lattice.forward()
    if check == True:
        ax = lattice.plot(xlim=(0,2), ylim = (0,2), zlim = (0,2))
        
interact(plot_func, 
        cell1_thetaz = widgets.FloatSlider(value=0.5,
                                               min=0,
                                               max=1,
                                               step=0.01),
        cell1_thetazP = widgets.FloatSlider(value=0.5,
                                               min=0,
                                               max=1,
                                               step=0.01),
        cell0_phiz = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell0_phiy = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell0_phizP = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell0_phiyP = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell1_phiz = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell1_phiy = widgets.FloatSlider(value=0.,
                                               min=-1,
                                               max=1,
                                               step=0.01),
        cell0_thetaz = widgets.FloatSlider(value=0.5,
                                               min=0,
                                               max=1,
                                               step=0.01),
        cell0_thetazP = widgets.FloatSlider(value=0.5,
                                               min=0,
                                               max=1,
                                               step=0.01),
        cell0_thetax = widgets.FloatSlider(value=0.,
                                               min=-0.5,
                                               max=0.5,
                                               step=0.01),
        cell0_thetaxP = widgets.FloatSlider(value=0.,
                                               min=-0.5,
                                               max=0.5,
                                               step=0.01),
        cell1_thetax = widgets.FloatSlider(value=0.,
                                               min=-0.5,
                                               max=0.5,
                                               step=0.001))
```
